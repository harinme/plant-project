```python
# yaml 파일을 학습이 가능하도록 경로 설정.
# key-value 데이터인 dict 데이터타입으로 data['train'], data['val'], data['nc'], data['names'] 에 넣어주는데,
# 가장 중요한 부분은 데이터 경로 설정임.
# 코랩에 맞는 경로로 바꾸어줌

import yaml

data = { 'train' : '/content/plant/train/images/',
         'val' : '/content/plant/valid/images/',
         'test' : '/content/plant/test/images',
         'names' :['미니-염자', '라울','프리티','레티지아'],
         'nc' : 4 }

# yaml파일 쓰기
with  open(  '/content/plant/plant.yaml',     'w'     )    as  f :
        yaml.dump( data, f   )

# 위에서 쓴 yaml파일 열기
with  open( '/content/plant/plant.yaml',   'r'   )   as   f :
       plant_yaml =  yaml.safe_load( f )
       display(  plant_yaml )
```

# Train (Python version)

```python
#  YOLOv8n 불러오기

from ultralytics import YOLO

#MS COCO Dataset 사전학습된 yolo8n모델을 로드함
# 다른 모델은 yolo8n 외에도 yolo8s, yolo8m, yolo8l, yolo8x등이 있다
# 그런데 뒤로 갈 수록 정확도는 높아지지만 용량과 처리속도가 느리다
 # load a pretrained YOLOv8n detection model
model=YOLO('yolov8n-seg.pt')
```

```python
# 원래 YOLOv8모델의 데이터 이름과 갯수 출력하기
print(type(model.names), len(model.names))
print(model.names)
```

```python
# 모델학습하기 data=yaml파일을 넣어준다.
model.train(data='/content/plant/plant.yaml', epochs=250, patience=30, batch=32, imgsz=640)
```
